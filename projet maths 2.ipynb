### Question 1

- $ \alpha $ est une constante correspondant au taux de reproduction des proies (indépendemment des prédateurs)
- $ \beta $ est le taux de mortalité des proies, il dépend donc du nombre de prédateurs
- $ \gamma $ est le taux de mortalité des prédateurs (il ne dépend pas du nombre de proies)
- $ \delta $ est le taux de reproduction des prédateurs qui dépend du nombre de proies


Notons $x = \begin{pmatrix} x_1 \\ x_2 \end{pmatrix}.$ On a alors : $\dot{x} = \begin{pmatrix} \dot{x_1} \\ \dot{x_2} \end{pmatrix} = f(x),$ où $f : \begin{pmatrix} x_1 \\ x_2 \end{pmatrix} \mapsto \begin{pmatrix} x_1(\alpha - \beta x_2) \\ -x_2(\gamma - \delta x_1) \end{pmatrix}$

On a alors $f\begin{pmatrix} x_1 \\ x_2 \end{pmatrix} = 0 \iff \begin{pmatrix} x_1 \\ x_2 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$ ou $\begin{pmatrix} x_1 \\ x_2 \end{pmatrix} = \begin{pmatrix} \frac{\gamma}{\delta} \\ \frac{\alpha}{\beta} \end{pmatrix}.$

$\boxed{Les\,points \,d'équilibre\, sont\, donc\, \begin{pmatrix} 0 \\ 0 \end{pmatrix} \,et\,  \bar{x} = \begin{pmatrix} \frac{\gamma}{\delta} \\ \frac{\alpha}{\beta} \end{pmatrix}.}$

### stabilité :

$J_f\begin{pmatrix} x_1 \\ x_2 \end{pmatrix} = \begin{pmatrix} \alpha - \beta x_2 & -\beta x_1 \\ \delta x_2 & \delta x_1 - \gamma \end{pmatrix}, $ d'où $J_f\begin{pmatrix} 0 \\ 0 \end{pmatrix} = \begin{pmatrix} \alpha & 0 \\ 0 & - \gamma \end{pmatrix}. $ Ses valeurs propres sont $ \alpha $ et $ - \gamma $, or $\alpha > 0$ donc $\boxed{\begin{pmatrix} 0 \\ 0 \end{pmatrix} est\,instable}$.

De la même manière, $J_f(\bar{x}) = \begin{pmatrix} 0 & - \frac{\beta \gamma}{\delta} \\ \frac{\delta \alpha}{\beta} & 0 \end{pmatrix}. $ Ses valeurs propres sont $ \pm i \sqrt{\alpha \gamma} ,$ on ne peut pas conclure.

### Question 2

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x1, x2, alpha=1, beta=1, gamma=1, delta=1):
    return (x1 * (alpha - beta * x2),-x2 * (gamma - delta * x1))

x1 = np.linspace(0,1,9)
x2 = np.linspace(0,5,9)
x1,x2=np.meshgrid(x1,x2)
y1,y2=f(x1,x2)
plt.quiver(x1,x2,y1,y2); #la fonction quiever nous permet de tracer un champ 2D de flèches 
                            #afin de visualiser le champ de vecteurs
```

```python
plt.streamplot(x1,x2,y1,y2); #on peut voir grace a la fonction streamplot le portrait de phase
        #on remarque bien que les courbes ne se coupent pas et qu'elles 'tournent' bien autour de x barre
```

Le champ de vecteurs (représenté en premier) met en évidence le comportement général des solutions et les deux points d'équilibre.Tandis que le portrait de phase (en deuxième) permet de réaliser que les solutions sont périodiques, bornées et à trajectoire fermée.

$ \begin{pmatrix} x_1 \\ x_2 \end{pmatrix} = \begin{pmatrix} 0 \\ x_{20} \exp({-\gamma t}) \end{pmatrix}$ serait alors une solution globale sur $\mathbb{R} \times \mathbb{R^+}^2.$

De plus, f est continue et sa différentielle partielle par rapport à x est définie et continue sur $\mathbb{R} \times \mathbb{R^+}^2$. Le théorème de Cauchy-Lipschitz assure donc l'unicité de la solution maximale.

D'après le théorème de domaine maximal d'existance, soit $ d((t, x(t)), \partial_{}(\mathbb{R} \times \mathbb{R^{+*}}^2)) \to 0 $ quand $t \to tm^-$, soit $\|x(t)\| \to + \infty$ quand $t \to tm^-$

Dans ce premier cas, on aurait $\|x(t)\| \to 0$ ou $\|x(t)\| \to + \infty$ quand $t \to tm^-$. (Le second cas est donc inclu dans le premier.)

```python
def solve_euler_explicit(f, x0, dt, t0, tf):
    t = [t0]
    x = [x0]
    a = t0
    while a < tf :
        x.append(x[-1] + dt * f(t[-1], x[-1]))
        a += dt 
        t.append(a)
    return t, x
```

```python
#on applique ce schema a l'equation y' = 2y - t dont les solutions sont calculables
def f(t, x):
    return 2*x - t
```

```python
t, x = solve_euler_explicit(f, 1, 0.001, 0, 0.5)
```

```python
plt.plot(t, x) #solution obtenue par Euler explicite
plt.plot(t, [3/4 * np.exp(2*i) + i/2 + 1/4 for i in t]); #solution exacte obtenue par calcul
```

Les deux solutions (exacte et obtenue par Euler explicite) sont bien superposées

On observe bien la convergence du schéma lorque $dt \to 0$

```python
L1 = []
L2 = []
M = []
H = [1 - 0.001*i for i in range(1000)] #on fait varier le temps delta.t(pour etudier la cvgce)
for k in H:
    maxi = 0 #maxi represente le max des xj approximé par Euler et des x(tj) (valeurs exactes)
    t, x = solve_euler_explicit(f, 1, k, 0, 0.5)
    for i,j in zip(t, x):
        if abs(3/4 * np.exp(2*i) + i/2 + 1/4 - j) > maxi :
            maxi = abs(3/4 * np.exp(2*i) + i/2 + 1/4 - j)
    L1.append(maxi/k)
    L2.append(maxi/(k*k))
    M.append(k)
```

```python
plt.plot(M, L1) #pour voir si l'ordre de convergence est >=1 ce qui est bien le cas d'apres
                #le graphe puisque maxi/(delta.t) est bien borné
```

```python
plt.plot(M, L2) #pour voir si l'ordre de convergence est >=2 ce qui n'est pas le cas d'apres
                #le graphe puisque maxi/(delta.t) diverge quand delta.t tend vers 0
```

On observe bien que $\frac{max(\|x^j - x(t_j)\|)}{\Delta t}$ est borné (en 0) alors que $\frac{max(\|x^j - x(t_j)\|)}{(\Delta t)^2}$ ne l'est pas.
$\boxed{L'ordre \,de \,convergence \,est \,donc\, bien \,1}$

#### Remarque (autre méthode)
Cette methode marche bien ici puisqu'on savait a l'avance que l'ordre de convergence n'est pas 'tres grand' (1,2,3 par exemple). Si nous n'avions aucune idée de l'ordre de convergence, une autre methode nous aurait permis d'approcher cet ordre p puis d'utiliser ce que nous venons de faire pour avoir un p exacte:
En effet si l'ordre de convergence est exactement p alors on peut dire qu'il existe une constante $c_v >0$ tel que $ {max(\|x^j - x(t_j)\|)}  \approx {c_v.(\Delta t)^p} $ pour ($\Delta t$ petit ) et en appliquant le logarithme on obtient $ \ln({max(\|x^j - x(t_j)\|)})  \approx { \ln{c_v} +p.\ln(\Delta t)} $ on trace alors $ \ln({max(\|x^j - x(t_j)\|)}) $ en fonction de $ \ln(\Delta t) $ et le coefficient directeur (pour les $\Delta t$ petit encore)  serait alors (a peu près) l'ordre de convergence 

```python
L = []
M = []
H = [0.1 - 0.001*i for i in range(100)] #on fait varier le temps delta.t(pour etudier la cvgce)
for k in H:
    maxi = 0 #maxi represente le max des xj approximé par Euler et des x(tj) (valeurs exactes)
    t, x = solve_euler_explicit(f, 1, k, 0, 0.5)
    for i,j in zip(t, x):
        if abs(3/4 * np.exp(2*i) + i/2 + 1/4 - j) > maxi :
            maxi = abs(3/4 * np.exp(2*i) + i/2 + 1/4 - j)
    L.append(np.log(maxi))
    M.append(np.log(k))
plt.plot(L,M)
plt.scatter(L,M);
```

```python
from scipy.optimize import newton
def solve_euler_implicit(f, x0, dt, t0, tf, itermax = 100):
    t = [t0]
    x = [x0]
    a = t0
    while a < tf :
        a += dt
        t.append(a)
        xsuiv = newton(lambda u: u - x[-1] - f(t[-1], u )* dt, x[-1], maxiter = itermax)
        x.append(xsuiv)
    return t, x
```

```python
#on applique euler implicite a la même equation y'=2y-t 
t, x = solve_euler_implicit(f, 1, 0.01, 0, 0.5)
plt.plot(t, x)#solution obtenue par Euler explicite
plt.plot(t, [3/4 * np.exp(2*i) + i/2 + 1/4 for i in t]);#solution exacte obtenue par calcul
```

Nous remarquons encore la quasi-superposition des deux solutions.

### Question 10

On a $H$ est dérivable (par rapport à t) on obtient alors en derivant:
$\frac{d(H(x_1,x_2) - H_0)}{dt} = -u_1(x_1, x_2)(H(x_1, x_2) - H_0)(\delta - \frac{\gamma}{x_1}) -u_2(x_1, x_2)(H(x_1, x_2) - H_0)(\beta - \frac{\alpha}{x_2})$.

Or en calculant le gradiant de H on trouve $\|\nabla H(x(t))\|^2 = (\delta - \frac{\gamma}{x_1})^2 + (\beta - \frac{\alpha}{x_2})^2$.

Si on choisit $ u_1(x_1, x_2) = k(\delta - \frac{\gamma}{x_1})$ et $u_2(x_1, x_2) = k(\beta - \frac{\alpha}{x_2})$, c'est à dire $\boxed{ u = k \nabla H}$, on obtient l'égalité voulue c'est à dire: 

$\frac{d(H(x_1,x_2) - H_0)}{dt} = -k. \|\nabla H(x(t))\|^2 .(H(x(t)) - H_0) $

Or x reste a une distance strictement positive de $\bar{x}$ donc il existe un $c'>0$ tel que  $0 < c' \leq \|x(t) - \bar{x}\| = (x_1 - \frac{\gamma}{\delta})^2 + (x_2 - \frac{\alpha}{\beta})^2 = (\frac{x_1}{\gamma})^2(\delta - \frac{\gamma}{x_1})^2 + (\frac{x_2}{\beta})^2(\beta - \frac{\alpha}{x_2})^2$ et on a $\|\nabla H(x(t))\|^2 = (\delta - \frac{\gamma}{x_1})^2 + (\beta - \frac{\alpha}{x_2})^2$.

Par stricte positivité de $x_1$ et $x_2$, il existe donc une constante $c$ telle que $\|\nabla H(x(t))\| \geq c > 0$.

En notant $m(t) = -k\|\nabla H(x(t))\|^2$ et $l(t) = H(x(t)) - H_0$, on a :

$\frac{dl}{dt} (t) = m(t)g(t)$. On obtient donc $l(t) = l_0 exp(\int_{t_0}^{t} m(s) ds)$ mais pour tout s entre $t_0$ et $t$, $m(s) \leq -k.c^2$ d'où $|l(t)| \leq |l(t_0)|e^{-k.c^2(t-t_0)}$.

Ainsi, $\boxed{H(x(t)) \,converge \,exponentiellement\,vers \,H_0\, lorsque\,
t\, tend \,vers \,l’infini\, si\, x\, reste \,à\, une \,distance \,strictement \,positive\, de \,\bar{x}}$
